```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
matches = pd.read_csv("matches.csv")
```

```python
matches[matches['winner'].isnull() == True]
```

```python
matches['winner'].fillna('No result',inplace=True)
```

```python
matches
```

```python
encoded_team = {
    'Mumbai Indians' : 1,
    'Chennai Super Kings' : 2,
    'Kolkata Knight Riders' : 3,
    'Royal Challengers Bangalore' : 4,
    'Kings XI Punjab'  : 5,
    'Rajasthan Royals'  : 6,
    'Delhi Daredevils' : 7,
    'Sunrisers Hyderabad' : 8,
    'Deccan Chargers' : 8,
    'Gujarat Lions' : 9,
    'Pune Warriors' : 10,
    'Delhi Capitals' : 7,
    'Rising Pune Supergiant' : 10,
    'Kochi Tuskers Kerala' :  11,
    'Rising Pune Supergiants' : 10,
    'No result' : 12
}

encoded_dict = {
    'team1' : encoded_team,
    'team2' : encoded_team,
    'toss_winner' : encoded_team,
    'winner' : encoded_team
}
matches.replace(encoded_dict,inplace=True)
```

```python
matches[matches['city'].isnull() == True]
```

```python
matches['city'].fillna('Dubai',inplace=True)
```

```python
matches[matches['city'].isnull() == True]
```

```python
from sklearn.preprocessing import LabelEncoder
ls = ['city','toss_decision','venue']
lbl_encoder = LabelEncoder()
for col in ls:
  matches[col] = lbl_encoder.fit_transform(matches[col])
  print(lbl_encoder.classes_)

matches
```

```python
x_train = matches[matches['Season']!='IPL-2019']
```

```python
x_train
```

```python
y_train = x_train['winner']
```

```python
y_train
```

```python
x_train = x_train[['team1' , 'team2' , 'city' , 'toss_decision', 'toss_winner' , 'venue' ]]
```

```python
x_train
```

```python
x_test = matches[matches['Season']=='IPL-2019']
x_test
```

```python
matches
```

```python
x_test
```

```python
y_test = x_test['winner']
```

```python
y_test
```

```python
x_test = x_test[['team1' , 'team2' , 'city' , 'toss_decision', 'toss_winner' , 'venue' ]]
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
logreg = LogisticRegression(penalty="none",max_iter=3500)
logreg.fit(x_train,y_train)
y_pred = logreg.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=1000)
rf.fit(x_train,y_train)
y_predict = rf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_predict))
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
for i in range (2,15):

  model = KNeighborsClassifier(n_neighbors=i)
  model.fit(x_train,y_train)
  predicted= model.predict(x_test)
  print(" K =  ",i , end = "  ")
  print("Accuracy:",metrics.accuracy_score(y_test, predicted))
confusion_matrix(y_test, y_pred)
```

```python
C=1
from sklearn import svm
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
svc = svm.SVC(kernel='linear',C=C).fit(x_train,y_train)
y_pred = svc.predict(x_test)
print("Linear SVM Acuracy:  ",accuracy_score(y_test,y_pred))
rbf_svc = svm.SVC(kernel='rbf',gamma=0.5,C=C).fit(x_train,y_train)
y_pred = rbf_svc.predict(x_test)
print("RBF SVM Acuracy:",accuracy_score(y_test,y_pred))
poly_svc = svm.SVC(kernel='poly',degree=6,C=C).fit(x_train,y_train)
y_pred = poly_svc.predict(x_test)
print("Polynmial SVM Acuracy:",accuracy_score(y_test,y_pred))
```
