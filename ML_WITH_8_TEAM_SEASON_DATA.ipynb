```python
matches[matches['winner'].isnull() == True]
matches['winner'].fillna('No result',inplace=True)
matches
```

```python
matches[matches['city'].isnull() == True]
matches['city'].fillna('Dubai',inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
ls = ['city','toss_decision','venue']
lbl_encoder = LabelEncoder()
for col in ls:
  matches[col] = lbl_encoder.fit_transform(matches[col])
```

```python
# Removing the seasons in which there are more than 8 teams on teams which are not consistent
```

```python
matches = matches[matches['Season']!='IPL-2010']
matches = matches[matches['Season']!='IPL-2011']
matches = matches[matches['Season']!='IPL-2012']
matches = matches[matches['Season']!='IPL-2013']
matches = matches[matches['Season']!='IPL-2016']
matches = matches[matches['Season']!='IPL-2017']
matches
```

```python
encoded_team = {
    'Mumbai Indians' : 1,
    'Chennai Super Kings' : 2,
    'Kolkata Knight Riders' : 3,
    'Royal Challengers Bangalore' : 4,
    'Kings XI Punjab'  : 5,
    'Rajasthan Royals'  : 6,
    'Delhi Daredevils' : 7,
    'Sunrisers Hyderabad' : 8,
    'Deccan Chargers' : 8,
    'Delhi Capitals' : 7,
    'No result' : 9
}

encoded_dict = {
    'team1' : encoded_team,
    'team2' : encoded_team,
    'toss_winner' : encoded_team,
    'winner' : encoded_team
}
matches.replace(encoded_dict,inplace=True)
```

```python
# Overall attempt
```

```python
testing_dataset = matches[matches['Season']=='IPL-2019']
training_dataset = matches[matches['Season']!='IPL-2019']
```

```python
train_x = training_dataset[['team1' , 'team2' , 'city' , 'toss_decision', 'toss_winner' , 'venue' ]]
train_y = training_dataset['winner']
test_x = testing_dataset[['team1' , 'team2' , 'city' , 'toss_decision', 'toss_winner' , 'venue' ]]
test_y = testing_dataset['winner']
train_x
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression(penalty="none",max_iter=5000)
logreg.fit(train_x,train_y)
y_predict = logreg.predict(test_x)
#print(y_predict)
print("Accuracy:",metrics.accuracy_score(test_y, y_predict))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, y_predict)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50)
rf.fit(train_x,train_y)
y_predict = rf.predict(test_x)
print("Accuracy:",metrics.accuracy_score(test_y, y_predict))
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, y_predict)
```

```python
from sklearn.neighbors import KNeighborsClassifier
for i in range (2,8):

  model = KNeighborsClassifier(n_neighbors=i)
  model.fit(train_x,train_y)
  predicted= model.predict(test_x)
  print("Accuracy:",metrics.accuracy_score(test_y, predicted))
  from sklearn.metrics import confusion_matrix
  print(confusion_matrix(test_y, y_predict))
```

```python
C=1
from sklearn import svm
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
svc = svm.SVC(kernel='linear',C=C).fit(train_x,train_y)
y_pred = svc.predict(test_x)
print(confusion_matrix(test_y, y_pred))
print("Linear SVM Acuracy:  ",accuracy_score(test_y,y_pred))
rbf_svc = svm.SVC(kernel='rbf',gamma=0.005,C=C).fit(train_x,train_y)
y_pred = rbf_svc.predict(test_x)
print(confusion_matrix(test_y, y_pred))
print("RBF SVM Acuracy:",accuracy_score(test_y,y_pred))
poly_svc = svm.SVC(kernel='poly',degree=3,C=C).fit(train_x,train_y)
y_pred = poly_svc.predict(test_x)
print("Polynmial SVM Acuracy:",accuracy_score(test_y,y_pred))
print(confusion_matrix(test_y, y_pred))
```
